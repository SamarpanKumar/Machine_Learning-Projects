```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(1,12,12)
print(x)
y = x**2 # Exponential power
print(y)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'#FF0000')
plt.show()

```

```python
plt.plot(x,y,'#F37726')
plt.xlabel('X Label').set_color('red')
plt.ylabel('Y Label').set_color('blue')
plt.title('x_y graph')
plt.show()
```

```python
fig=plt.figure()
axes=fig.add_axes ([0.1,0.2,1.3,1.4])
#([0.1,0.1,0.8,0.8])
axes.plot(x,y)
```

```python
fig =plt.figure()
axes=fig.add_axes([0.1,0.1,1.3,0.5])
axes.plot(x,y)

axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
plt.show()
```

```python
fig = plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.4,0.2,0.4,0.4])
```

```python
fig=plt.figure()

axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.45,0.15,0.4,0.3])
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.45,0.15,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('Larger plot')
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes=plt.subplots()
axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows =3,ncols=3)
axes[0][2].plot(x,y)
axes[1][1].plot(y,x)
```

```python
fig,axes=plt.subplots(nrows=3,ncols=3)
axes[0][2].plot(x,y)
axes[1][0].plot(y,x)
axes[2][1].plot(x,x)
```

```python
fig,axes=plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows =2,ncols=2)
axes[0][0].plot(x,y)
axes[0][0].set_title("First")
axes[1][0].plot(x,y)
axes[1][0].set_title('Second')
plt.tight_layout()
```
